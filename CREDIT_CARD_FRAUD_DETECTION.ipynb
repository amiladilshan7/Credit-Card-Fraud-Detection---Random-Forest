Above Dataset is highly imbalanced, with only 492 fraud cases (Class 1) out of 284,807 total transactions.

This imbalance can cause issues when training machine learning models, as they may become biased toward the majority class (non-fraud). Here's how you can preprocess and handle this imbalance:

## 1. Understand the Imbalance

Class Distribution:

1. Non-fraud (Class 0): 284,315

2. Fraud (Class 1): 492

Imbalance Ratio: ~578:1 (very severe imbalance)

### **So we need to Handle the Imbalanced data**

Before addressing the imbalance, split the dataset into training and testing sets to avoid data leakage:

```python
# seperating the data for analysis

# Separate features (X) and target (y)
legit = credit_card_data.drop("Class", axis=1)
fraud = credit_card_data["Class"]

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(legit, fraud, test_size=0.2, random_state=42, stratify=fraud)
#Why stratify= fraud? Ensures the same class distribution is maintained in both training and testing sets.

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

```python
# statical measures of the data
legit.Amount.describe()
```

 ## Handle Imbalanced Data

There are several techniques to handle imbalanced datasets.

1.   Oversampling the Minority Class: Use techniques like SMOTE (Synthetic Minority Oversampling Technique) to generate synthetic samples for the minority class.


2.   Undersampling the Majority Class: Randomly remove samples from the majority class to balance the dataset.




```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=42)
X_train_res, y_train_res = undersampler.fit_resample(X_train, y_train)

print("Resampled training set shape:", X_train_res.shape)
print("Resampled class distribution:\n", y_train_res.value_counts())
```

Concatanating 2 Dataframes

```python
new_dataset = pd.concat([X_train_res, y_train_res], axis=1)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Split the Concatenated Dataset

Since new_dataset contains both features and the target, we need to split it back into X (features) and y (target) for training:



```python
# Separate features (X) and target (y)
X_train_balanced = new_dataset.drop("Class", axis=1)  # Features
y_train_balanced = new_dataset["Class"]  # Target

print("Balanced training set shape:", X_train_balanced)
print("Balanced class distribution:\n", y_train_balanced)
```

## Train a Model

Now, train a model on the balanced dataset. Let's use a Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
model = RandomForestClassifier(random_state=42)

# Train the model on the balanced data
model.fit(X_train_balanced, y_train_balanced)
```

## Evaluate the Model on the Test Set


Evaluate the model on the original (unresampled) test set (X_test, y_test) to ensure the model's performance is tested on realistic, imbalanced data.

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

 ### Evaluate Using Metrics for Imbalanced Data

Use metrics like Confusion Matrix, Precision, Recall, F1-Score, and ROC-AUC to evaluate the model:

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification Report (Precision, Recall, F1-Score)
print("Classification Report:\n", classification_report(y_test, y_pred))

# ROC-AUC Score
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred))
```

### Accuracy score

```python
X_train_prediction = model.predict(X_train_balanced)
training_data_accuracy = accuracy_score(X_train_prediction, y_train_balanced)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```
